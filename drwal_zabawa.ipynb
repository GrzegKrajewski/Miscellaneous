```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cena=[11.90,11.90,12.90,13.60,14.60,14.60,15,16.90,17.40,23.90,28.90]
rok=[2012,2013,2014,2015,2017,2018,2019,2020,2021,2022,2023]
inflacja=[0,0.9,0,-0.9,-0.6,2,1.6,2.3,3.4,5.1,14.4,8]
data = {'rok':rok ,'cena':cena}

df=pd.DataFrame(data)
```

```python
sns.lineplot(df,x='rok',y='cena',color='red')


for i, j in zip(df['rok'], df['cena']):
    plt.text(i, j, f'{j}', va='top',color='red',fontsize=14)
    plt.scatter(i, j, marker='o',color='red')

plt.title('Cena Kanapki Drwala na przestrzeni lat',size=22)
plt.ylim(0,30)
plt.xlim(2011,2024)
plt.ticklabel_format(style='plain', axis='y')
plt.style.use('Solarize_Light2')
```
